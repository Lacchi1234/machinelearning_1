```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/Housing.csv")
df.head()
```

```python
df=pd.get_dummies(df)
df.head()
```

this converts  any string into binary for machine to learn and duplicates...this process is called binary encode

```python
df.drop(['mainroad_no','guestroom_no','basement_yes','hotwaterheating_yes','airconditioning_yes'],axis=1,inplace=True)
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
y_pred=lm.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

print("MSE          ---- >" , mean_squared_error(y_test,y_pred))
print("RMSE       ----- > " , np.sqrt(mean_squared_error(y_test,y_pred)))
print ("MAE        ----- >", mean_absolute_error(y_test,y_pred))
print("r2 score “ ----- >",r2_score(y_test,y_pred))
```
