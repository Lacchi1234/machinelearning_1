```python
q1=data['Age'].quantile(0.25)
q3=data['Age'].quantile(0.75)
IQR=q3-q1
lower=q1-(1.5*IQR)
upper=q3+(1.5*IQR)
data['Age']=np.where(data['Age']>upper,upper,np.where(data['Age']<lower,lower,data['Age']))
sns.boxplot(data['Age'])
```
